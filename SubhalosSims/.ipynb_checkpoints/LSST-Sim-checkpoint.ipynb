```python
def generate_stars(density=5,xlims=[-20, 20],ylims=[-20, 20]):
    area = (xlims[1] - xlims[0])*(ylims[1] - ylims[0])
    n_stars = np.random.poisson(area*density)
    print("Generating", n_stars, "stars")
    x_coords = (xlims[1]-xlims[0])*np.random.random_sample(n_stars) + xlims[0]
    y_coords = (ylims[1]-ylims[0])*np.random.random_sample(n_stars) + ylims[0]
    return x_coords, y_coords
```

```python
xlims = [-20, 20]
x_coords, y_coords = generate_stars(density=2)
im, xedges, yedges = np.histogram2d(x_coords, y_coords, bins=40, range=[[-20,20],[-20,20]])
```

```python
# Point lens properties
G = 2
M_l = 2 # Lens mass
D_l = 2 # Lens distance

D_s = 10 # Fix all sources at same distance
```

```python
def deflection_theta_ps(pos_s, pos_l):
    b_sl_vec = np.array(pos_s) - np.array(pos_l)
    b_sl_norm = np.linalg.norm(b_sl_vec)
    b_sl_unit = b_sl_vec/b_sl_norm
    return -(1-D_l/D_s)*(4*G*M_l/b_sl_norm)*b_sl_unit
```

```python
poss_s = np.transpose([x_coords, y_coords])
```

```python
n_lens = 6
poss_l = []

for i_lens in range(n_lens):
    pos_l = (xlims[1]-xlims[0])*np.random.random_sample(2) + xlims[0]
    poss_l.append(pos_l)
    
vmin = 1
vmax = 6
t_tot = 10
n_obs = 2
vel_l = np.array([(vmax-vmin)*np.random.random_sample(2) + vmin for i in range(n_lens)])


for i_obs in range(n_obs):
    
    deflections = []
    deflection_tot = np.zeros(np.shape(np.array(poss_s)))
    
    for i_lens in range(n_lens):
        deflection = np.array([deflection_theta_ps(pos_s, poss_l[i_lens] + i_obs*vel_l[i_lens] ) for pos_s in poss_s])
        deflections.append(deflection)
        deflection_tot += deflection

    x_coords_lensed, y_coords_lensed = np.transpose(poss_s + deflection_tot)

    im_lensed, xedges, yedges = np.histogram2d(x_coords_lensed, y_coords_lensed, bins=40, range=[[-20,20],[-20,20]])

    # Make plot with vertical (default) colorbar
    fig, ax = plt.subplots()
    cax = ax.pcolor(xedges, yedges, (im_lensed - im).T, vmin=-50, vmax=50)
    ax.set_title('Lensed minus unlensed')
    for i_lens in range(n_lens): 
        x_l, y_l = poss_l[i_lens] + i_obs*vel_l[i_lens] 
        ax.scatter(x_l, y_l)
    # Add colorbar
    cbar = fig.colorbar(cax, ticks=np.linspace(-50,50,10))

    plt.xlim(-20,20)
    plt.ylim(-20,20)
    plt.ioff()
```
