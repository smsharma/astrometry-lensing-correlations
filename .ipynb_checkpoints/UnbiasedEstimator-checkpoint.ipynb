```python
from scipy.special import lpn
```

```python
def Cl_to_wtheta(C_ells, theta=0):
    ells = np.arange(len(C_ells))
    return (1/(4*np.pi))*np.sum((2*ells + 1)*C_ells*lpn(ells[-1], np.cos(theta*np.pi/180.))[0])
```

```python
thetas = np.linspace(0,10, 100)
wthetas = [Cl_to_wtheta(hp.anafast(ps_map/np.mean(ps_map)), theta = theta) for theta in thetas]
wthetas_exp = [Cl_to_wtheta(hp.anafast(exposure/np.mean(exposure)), theta = theta) for theta in thetas]
```

```python
plt.plot(thetas, wthetas)
plt.plot(thetas, wthetas_exp)
```
